```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# matplotib은 시각화에 사용
```

```python
irisdata = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning\
-databases/iris/iris.data', header = None) # header = None : column 이름 지정 없이 read(없을 경우 첫 줄의 데이터를 이름으로 인식)
irisdata.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Name']
irisdata.head(10) # 위에서부터 10개만 출력
```

```python
len(irisdata) # 행의 개수 출력
```

```python
groups = irisdata.groupby('Name')
# groupby는 기준에 따라 데이터를 그룹화
list(groups)[1]
# 0은 setosa, 1은 versicolor, 2는 virginica
```

```python
iris = pd.read_csv('data/iris.csv')
# 저장한 데이터를 불러올 때는 sub directory 이름만 파일 앞에 넣어주면 된다.
iris.head() # 기본값이 5 row
```

4. Data visualization : matplotlib

```python
plt.figure(figsize = (6,4))
# figsize는 시각화할 크기 설정
plt.subplot(2,1,1)
# subplot을 통해 시각화할 데이터들의 좌표를 입력(가로, 세로, 위치)
plt.subplot(2,1,2)
plt.show()
# show는 작업의 종료를 의미(필수)
```

```python
iris.boxplot()
plt.show()
```

```python
iris.hist()
plt.show()
```

```python
iris.plot()
plt.show()
```

```python
plt.scatter(iris['PetalL'], iris['PetalW'])
# scatter는 x축과 y축 2개의 변수가 필요
plt.show()
```

```python
year, number = (['1st Year', '2nd Year', '3rd Year', '4th Year'], [14, 11, 15, 18])
plt.figure(figsize = (12,3))
plt.subplot(1,2,1)
plt.bar(year, number)
plt.subplot(1,2,2)
plt.barh(year, number)
plt.show()
```

```python
students = pd.Series([14, 11, 15, 18], index = ['1st Year', '2nd Year', '3rd Year', '4th Year'])
print(students)
plt.figure()
# default size가 할당
students.sort_values().plot(kind = 'barh', grid = True, figsize = (6,3))
plt.show()
```

```python
setosa = iris.iloc[0:50,:]
versicolor = iris.iloc[50:100,:]
verginica = iris.iloc[100:150,:]
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

fig, ax = plt.subplots()
# fig는 도화지(캔버스)의 역할
ax.plot(x, y)
ax.set_title('Simple plot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(iris['SepalL'])
plt.hist(setosa['SepalL'])
plt.hist(versicolor['SepalL'])
plt.hist(verginica['SepalL'])
# 순서는 입력한 순서대로 위에 덧붙인다.
plt.xlabel('Sepal Length(cm)', fontsize = 10)
plt.ylabel('Frequency', fontsize = 10)
plt.show()
```

```python
kinds = [setosa, versicolor, verginica]
names = ['Setosa', 'Versicolor', 'Verginica']

fig, ax = plt.subplots(figsize = (6,4))
plt.hist(iris['SepalL'])
for x, y in zip(names, kinds):
    ax.hist(y.SepalL, label = x)
plt.title('Distribution of Sepal by Kinds')
plt.xlabel('Sepal Length(cm)', fontsize = 10)
ax.legend(fontsize = 10, loc='upper right')
# legend는 범주를 만드는 함수, loc는 범주의 위치 지정
plt.ylabel('Frequency', fontsize = 10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,5))
for name, kind in zip(names, kinds):
    ax.plot(kind.PetalL,          # x축 정보
            kind.PetalW,          # y축 정보
            marker = '*',         # 표현 모양 설정
            linestyle = '',       # 선의 삭제
            label = name)         # 각 구분별로 레이블링
ax.legend(fontsize = 10, loc = 'upper left')
fig.suptitle('Distribution of Iris by its Kinds')
# 전체 도표의 타이틀(plt.title과 동일)
plt.xlabel('Petal Length(cm)', fontsize = 10)
plt.ylabel('Petal Width(cm)', fontsize = 10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.pie(students, explode = (0,0,0,0.1), labels = students.index,
       autopct = '%1.2f%%', shadow = True, startangle = 90)
# pie차트 작성, autopct는 row data를 비율로 보여준다.('% %'는 표기 방식)
# explode는 차트를 떨어뜨려 표현 가능
fig.suptitle("Students by Year", fontsize = 14)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')   # 배경 스타일 지정(white, dark, ticks, etc.)
iris_scat = sns.scatterplot(x='PetalL', y='PetalW', palette='bright',
                            hue='Name', style='Name', s=80, data = iris)
iris_scat.set_title('Iris Distribution by Sepal', fontsize = 14)
iris_scat.set(xlabel = 'Sepal Length(cm)', ylabel = 'Sepal Width(cm)')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='Name',y='SepalW', data=iris)
plt.show()
```
